```python
import requests
from bs4 import BeautifulSoup

HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
BASEURL = "https://247sports.com"

players = []

for i in range(1,44):
    request = requests.get(f"{BASEURL}/Season/2024-Football/CompositeRecruitRankings/?ViewPath=~%2FViews%2FSkyNet%2FPlayerSportRanking%2F_SimpleSetForSeason.ascx&InstitutionGroup=highschool&Page={i}", headers=HEADERS)
    soup = BeautifulSoup(request.content, "html.parser")
    recruits = soup.find_all("li", class_="rankings-page__list-item")
    for recruit in recruits:
            if recruit.text.strip() == "Load More":
                continue
            name = recruit.find("a", class_="rankings-page__name-link", href=True)
            position = recruit.find("div", class_="position")
            home = recruit.find("span", class_="meta")
            metrics = recruit.find("div", class_="metrics")

            committment_status = recruit.find("div", class_="status")
            if committment_status.find("div", class_="rankings-page__crystal-ball"):
                college_committed_to = "N/A"
            else:
                try:
                    college = committment_status.find("img", class_="jsonly", alt=True)
                    college_committed_to = college["alt"]
                except:
                    college_committed_to = "N/A"
            
            height_weight = metrics.text.strip().split("/")
            home_town = home.text.strip().split("\n")
            high_school = home_town[0].strip()
            town_state = home_town[1].strip().replace("(", "").replace(")","")

            players.append({"name":name.text.strip(), "link": name["href"], "position":position.text.strip(), "high_school": high_school, "hometown": town_state, "height":height_weight[0].strip(), "weight":height_weight[1].strip(), "college_committed": college_committed_to})
```
